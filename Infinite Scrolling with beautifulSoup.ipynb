```python
data = open('DS.html',encoding='utf-8')
```

```python
doc = BeautifulSoup(data, "html.parser")
```

```python
doc2 = BeautifulSoup(doc.prettify(),"html.parser")
```

```python
print(doc2)
```

```python
spans = doc2.find_all('span', {'class' : 'title'})
names = [span.get_text() for span in spans]
ls_names = []
for name in names:
    ls_names.append(name.strip())
```

```python
divs = doc2.find_all('div', {'class' : ['col search_price responsive_secondrow','col search_price discounted responsive_secondrow']})
prices = [div.get_text() for div in divs]
ls_prices = []
for price in prices:
    ls_prices.append(price.strip())
```

```python
divrr = doc2.find_all('div', {'class' : 'col search_released responsive_secondrow'})
release_date = [div.get_text() for div in divrr]
ls_release = []
for release in release_date:
    ls_release.append(release.strip())
```

```python
revs = doc2.find_all('span', {'class' : ['search_review_summary positive','search_review_summary mixed','search_review_summary negative']})
```

```python
print(len(ls_names))
print(len(ls_prices))
print(len(ls_release))
print(len(revs))
```

```python
trying = doc2.find_all('div',{'class':"col search_reviewscore responsive_secondrow"})
```

```python
ls = []
for i in trying:
    n = str(i).split("<")[2].split(";")
    ls.append(n)
```

```python
rev_list = []
for i in range(550):
    if len(ls[i])>2:
        rev_list.append(ls[i][2])
    else:
        rev_list.append('No user reviews yet')
```

```python
rating_rev = []
for j in rev_list:
    rating_rev.append((j[:3]))
```

```python
user_rev = []
for j in range(len(rev_list)):
    user_rev.append(rev_list[j].split(" ")[3])
```

```python
link = doc2.find_all('a',{'class':'search_result_row ds_collapse_flag'})[0]['href']
link
```

```python
steamlinks = []
for l in range(len(ls_names)):
    steamlinks.append(doc2.find_all('a',{'class':['search_result_row ds_collapse_flag app_impression_tracked','search_result_row ds_collapse_flag']})[l]['href'])
steamlinks
```

```python
platforms = doc2.find_all('p')
plat = []
for i in range(len(platforms)):
    for j in range(len(platforms[i].find_all('span'))):
        plat.append(platforms[i].find_all('span')[j]['class'])
plat
```

```python
plats = pd.DataFrame(list(zip(plat)),columns=['Platform'])
plats.head()
```

```python
plats['Counter'] = 1
count = 0
for i in range(len(plats['Platform'])):
    if plats['Platform'][i]==['platform_img', 'win']:
        plats['Counter'][i] = count + 1
        count = count + 1
    elif plats['Platform'][i]==['platform_img', 'mac']:
        plats['Counter'][i] = count
    elif plats['Platform'][i]==['platform_img', 'linux']:
        plats['Counter'][i] = count
    elif plats['Platform'][i]==['vr_supported']:
        plats['Counter'][i] = count
    else:
        pass
```

```python
plats.head()
```

```python
ls_pl = []
for sep in range(len(plats['Platform'])):
    if len(plats['Platform'][sep])==2:
        ls_pl.append(plats['Platform'][sep][1])
    else:
        ls_pl.append(plats['Platform'][sep][0])
```

```python
plats['Platform_Supported'] = ls_pl
```

```python
groups = plats.groupby('Counter').agg([('Platform_Supported',','.join)])
plats['Platforms'] = groups['Platform_Supported']
ls_plat = groups['Platform_Supported']
```

```python
ls_plat['Platform_Supported']
```

```python
games = pd.DataFrame(list(zip(ls_names,ls_prices,ls_release,rating_rev,user_rev,steamlinks,ls_plat['Platform_Supported'])),
                     columns=['Title','Price','Release Date','Game Rating','Users Reviewed','Steam Link','Platforms Supported'])
```

```python
games['len>12'] = games['Price'].str.len()>12
```

```python
for a in range(len(ls_names)):
    if games['len>12'][a]==True:
        games['Price'][a] = games['Price'][a].strip()[40:]
```

```python
games.drop(columns=['len>12'],inplace=True)
```
